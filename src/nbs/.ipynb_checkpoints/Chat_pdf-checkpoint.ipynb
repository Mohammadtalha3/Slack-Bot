```python
from sentence_transformers import SentenceTransformer

def get_embeddings(text_list):
    """
    This function takes a list of text and returns the corresponding embeddings using the all-MiniLM-L6-v2 model.
    
    Args:
        text_list (list of str): List of sentences or paragraphs to embed.
    
    Returns:
        embeddings (list of np.array): List of embeddings for each input text.
    """
    # Load the MiniLM-L6-v2 model
    model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
    
    # Generate embeddings for the list of texts
    embeddings = model.encode(text_list, convert_to_numpy=True)
    
    return embeddings
```

```python
def retrieve_documents(query, limit=5):
    query_embedding = get_embeddings(query)  # Embed the query
    class_name = 'Documents'
    results = client.query.get(class_name, properties=["text", "title"]).with_near_vector({
        "vector": query_embedding,
        "certainty": 0.7  # Adjust as needed
    }).with_limit(limit).do()
    
    return results['data']['Get'][class_name]
```

```python
query_text = "Forms in django"
retrieved_docs = retrieve_documents(query_text)

print("Retrieved Documents:")
for doc in retrieved_docs:
    print(f"Text: {doc['text'][:100]}...") 
    # print(doc)
```

```python
def retrieve_documents(query, limit=5):
    # Set the class name where your documents are stored
    class_name = 'Documents'
    
    # Get the query embedding
    query_embedding = get_embeddings(query)  # Embed the query
    
    try:
        # Query Weaviate for documents near the query vector
        results = client.query.get(class_name, properties=["text"]).with_near_vector({
            "vector": query_embedding,
            "certainty": 0.5    # Adjust as needed
        }).with_limit(limit).do()

        # Print the full response to inspect its structure
        print("Weaviate Response:", results)

        # Extract and return documents if 'data' exists
        return results.get('data', {}).get('Get', {}).get(class_name, [])
    
    except Exception as e:
        print(f"Error during query: {e}")
        return []

# Example usage:
query_text = "Forms in django"
retrieved_docs = retrieve_documents(query_text)

# Print retrieved documents
print("Retrieved Documents:")
for doc in retrieved_docs:
    print(doc)
```
